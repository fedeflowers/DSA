the idea is to generate all the strings.
how does it works:
base case: when it creates the final string it returns it.
recursion: if the pattern contains I than the el to put must be greater than the previous. else if contains a D is the opposite.
it works because the nums list is ordered so the number taken is always the smallest.
if it's not possible to create a string None is returned so discard all Nones. and keep the first string created with smallest numbers.
it is also greedy because the smallest is always taken
it is not necessary to use numbers since with strings "2" > "1" and so on...

```python
class Solution:
    def smallestNumber(self, pattern: str) -> str:
        nums = [str(i) for i in range(1, 10)]

        def build(start, pattern, index, nums):
            if index == len(pattern):
                return start

            for i in range(len(nums)):
                if pattern[index] == 'I':
                    if nums[i] > start[index]:
                        res = build(start + nums[i], pattern, index+1, nums[:i] + nums[i+1:])
                        if res:
                            return res
                else:
                    if nums[i] < start[index]:
                        res = build(start + nums[i], pattern, index+1, nums[:i] + nums[i+1:])
                        if res:
                            return res
        
        for i in range(len(nums)):
            res = build(nums[i], pattern, 0, nums[:i] + nums[i+1:])
            if res:
                return res
```

# ðŸ§  Leetcode 51 â€“ N-Queens

## ðŸ” Approach: Backtracking

### ðŸ’¡ Intuition

We aim to generate valid board states **without placing a queen where another can attack it**. This makes it a perfect candidate for **backtracking**.

At each step:
- Place a queen in a safe position.
- Recurse to the next row.
- If we can't place a queen, backtrack and try the next column.

A queen is **under attack** if another queen is in the same:
- Row
- Column
- Diagonal
- Anti-diagonal

To backtrack:
- Implement a recursive function that tries to place a queen row-by-row.
- Maintain sets for used columns, diagonals, and anti-diagonals.
- If a placement is valid, add it to the board and continue.
- After recursion, remove the queen and revert the sets.

---

## âš™ï¸ Strategy

### Encode State:
Each recursive `backtrack(row)` call:
- Tries to place a queen in `row`
- Skips columns, diagonals, or anti-diagonals already used

Track used indices with:
- `cols` for used columns
- `diagonals` for `row - col`
- `antiDiagonals` for `row + col`

If `row == n`, we've placed all queens â€” store the board.

---

## ðŸ”¢ Diagonals Explained

- **Main Diagonal (â†˜)**: `row - col` is constant
- **Anti-Diagonal (â†™)**: `row + col` is constant

---

## ðŸ§® Algorithm

1. Define `backtrack(row, diagonals, antiDiagonals, cols, state)`
2. If `row == n`, append a deep copy of `state` to the result
3. For each column:
   - Skip if column, diagonal, or anti-diagonal is already used
   - Place a queen at `(row, col)`
   - Update sets and board
   - Recurse to `row + 1`
   - Backtrack (remove queen and update sets)

---

## ðŸ§  Time & Space Complexity

Let `N` be the size of the board:

### Time Complexity: `O(N!)`
- First row: `N` options
- Second row: `â‰¤ N - 2` (due to diagonals and column restrictions)
- Third row: `â‰¤ N - 4`, and so on...
- Total time: `O(N!)`

### Space Complexity: `O(N^2)`
- Sets: `O(N)`


```python
# class Solution:
#     def solveNQueens(self, n: int) -> List[List[str]]:
        #i'm checking all positions but it's too slow, i should just check rows
        # #is the board valid
        # def valid(board):
        #     '''
        #     board = nxn board
        #     a board it's valid if the queens do not attack each other
        #     '''
        #     #if there is a queen check row, col and diagonals
        #     for i in range(n):
        #         for j in range(n):
        #             if board[i][j] == 'Q':
        #                 #check row
        #                 for k in range(0, n):
        #                     if k != j:
        #                         if board[i][k] == 'Q':
        #                             return False
        #                 #check col
        #                 for k in range(0, n):
        #                     if i != k:
        #                         if board[k][j] == 'Q':
        #                             return False

        #                 #check diagonals
        #                 diags = [(-1,-1), (1,1), (-1,1), (1, -1)]
        #                 for d in diags:
        #                     temp_i = i
        #                     temp_j = j
        #                     while 0 <= temp_i < n and 0 <= temp_j <n:
        #                         x, y = d
        #                         if temp_i != i and temp_j != j and board[temp_i][temp_j] == 'Q':
        #                             return False
        #                         temp_i -= x
        #                         temp_j -=y
        #     return True 

        # #try all possible combinations until n == 0 and valid
        # base = [['.'] * n for _ in range(n)]
        # res = []
        # def backtrack(queens, board, row):
        #     if queens == 0:
        #         res.append(["".join(r) for r in board])
        #         return

        #     for i in range(row, n):  # row-level pruning
        #         for j in range(n):
        #             board[i][j] = "Q"
        #             if valid(board):
        #                 backtrack(queens - 1, board, i + 1)  # move to next row
        #             board[i][j] = "."  # backtrack

        # backtrack(n, base, 0)
        # return res


class Solution:
    def solveNQueens(self, n):
        # Making use of a helper function to get the
        # solutions in the correct output format
        def create_board(state):
            board = []
            for row in state:
                board.append("".join(row))
            return board

        def backtrack(row, diagonals, anti_diagonals, cols, state):
            # Base case - N queens have been placed
            if row == n:
                ans.append(create_board(state))
                return

            for col in range(n):
                curr_diagonal = row - col
                curr_anti_diagonal = row + col
                # If the queen is not placeable
                if (
                    col in cols
                    or curr_diagonal in diagonals
                    or curr_anti_diagonal in anti_diagonals
                ):
                    continue

                # "Add" the queen to the board
                cols.add(col)
                diagonals.add(curr_diagonal)
                anti_diagonals.add(curr_anti_diagonal)
                state[row][col] = "Q"

                # Move on to the next row with the updated board state
                backtrack(row + 1, diagonals, anti_diagonals, cols, state)

                # "Remove" the queen from the board since we have already
                # explored all valid paths using the above function call
                cols.remove(col)
                diagonals.remove(curr_diagonal)
                anti_diagonals.remove(curr_anti_diagonal)
                state[row][col] = "."

        ans = []
        empty_board = [["."] * n for _ in range(n)]
        backtrack(0, set(), set(), set(), empty_board)
        return ans
```

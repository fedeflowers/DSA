# 3174. Clear Digits

Use a stack: append if the char is not a digit, pop if it's a digit and there is something in the stack.

```python
class Solution:
    def clearDigits(self, s: str) -> str:
        stack = []
        for c in s:
            if c.isdigit():
                if stack:
                    stack.pop()
            else:
                stack.append(c)
        return "".join(stack)
```

Given N as input, count the number of patterned strings.

Patterned strings are strings whose substring gives the string itself of length N when repeated any number of times. (Only A and B characters are present)

Input: 4
Output: 4

The strings are {AAAA, BBBB, ABAB, BABA}

{AABA, AAAB, AABB} are not valid.

```python
def patterned_strings(n):
    primitive_counts = [0] * (n+1)
    all_counts = [0]
    for i in range(1, n+1):
        all_counts.append(2**i)
    
    for i in range(len(all_counts)):
        primitive = 0
        for j in range(1, i):
            if i % j == 0:
                primitive += primitive_counts[j]

        primitive_counts[i] = all_counts[i] - primitive

    res = 0
    for i in range(1, n):
        if n % i == 0:
            res += primitive_counts[i]
    return res

patterned_strings(6)
```

Consider a social network with n users. You have been given vector<vector< int >> friends. where friends[i][0] and friends[i][1] are friends. <br>
 A person 'a' can view profile of 'b' if he is friend or has a indirect common friend. Find out how many profiles can each user view.<br>
Constraint :<br>
n <= 10 ^ 5<br>
friends.size() <= n<br>
E.g. :<br>
n = 7<br>
friends = [{1,2}, {2,3}, {3,4}, {5,6}, {6,7}]<br>
result = [4,4,4,4,3,3]<br>

```python
from collections import defaultdict, deque
#or USE DSU
def profile_views(n, friends):
    graph = defaultdict(list)
    
    for u, v in friends:
        graph[u].append(v)
        graph[v].append(u)

    visited = [False] * (n + 1)
    result = [0] * (n + 1)

    for i in range(1, n + 1):
        if not visited[i]:
            # BFS to find the component
            queue = deque([i])
            visited[i] = True
            component = [i]

            while queue:
                node = queue.popleft()
                for neighbor in graph[node]:
                    if not visited[neighbor]:
                        visited[neighbor] = True
                        queue.append(neighbor)
                        component.append(neighbor)

            size = len(component)
            for user in component:
                result[user] = size

    return result[1:]  # exclude index 0 if 1-indexed

n = 7
friends = [[1,2], [2,3], [3,4], [5,6], [6,7]]
profile_views(n, friends)
```

You are given a class LongestConsecutiveSequence which exposes a function addNumberAndCompute that accepts a number as an argument. <br>
This function can be called by the client any number of times as per their liking. After each computation the function is supposed<br>
to return the length of the longest consecutive sequence you have seen till now.<br>
addNumberAndCompute(1), returns 1<br>
addNumberAndCompute(2), returns 2<br>
addNumberAndCompute(4), returns 2<br>
addNumberAndCompute(3), return 4<br>

I guess sequence means 1 to n e.g first 3 is 2 because 4 is not in the sequence, we need 3 to get to the sequence
if i ahve 4,5,6,7 is that the longest sequence?
maybe DSU ?

```python
class DSU:
    def __init__(self):
        self.parent = {}
        self.size = {}
        self.max_seq = 0

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]

    def union(self, x, y):
        px, py = self.find(x), self.find(y)
        if px == py:
            return
        # Merge smaller set into larger
        if self.size[px] < self.size[py]:
            px, py = py, px
        self.parent[py] = px
        self.size[px] += self.size[py]

    def add(self, num):
        if num in self.parent:
            return self.max_seq

        self.parent[num] = num
        self.size[num] = 1

        # Try merging with left and right neighbors
        if num - 1 in self.parent:
            self.union(num, num - 1)
        if num + 1 in self.parent:
            self.union(num, num + 1)

        root = self.find(num)
        self.max_seq = max(self.max_seq, self.size[root])
        return self.max_seq


class LongestConsecutiveSequence:
    def __init__(self):
        self.dsu = DSU()

    def addNumberAndCompute(self, num):
        return self.dsu.add(num)


# Test
lcs = LongestConsecutiveSequence()
print(lcs.addNumberAndCompute(1))  # 1
print(lcs.addNumberAndCompute(2))  # 2
print(lcs.addNumberAndCompute(5))  # 2
print(lcs.addNumberAndCompute(4))  # 2
print(lcs.addNumberAndCompute(3))  # 5 (1-5 connected)
print(lcs.addNumberAndCompute(7))  
print(lcs.addNumberAndCompute(6))  
print(lcs.addNumberAndCompute(-1))  
print(lcs.addNumberAndCompute(0))  



```

8) Given a binary tree, return true if all the values of node are equal
Given a binary tree, return maximum binary sub tree size where all values of node are equal

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def largestUnivalueSubtree(self, root: TreeNode) -> int:
        self.max_size = 0 # To store the maximum univalue subtree size found

        # Helper function: returns (is_univalued, subtree_size, value)
        def dfs(node):
            if not node:
                return (True, 0, None) # An empty tree is considered univalued of size 0

            is_left_univalued, left_size, left_val = dfs(node.left)
            is_right_univalued, right_size, right_val = dfs(node.right)

            current_is_univalued = True
            current_size = 1 # Current node itself

            # Check if left subtree is univalued and matches current node's value
            if not is_left_univalued or (node.left and node.val != left_val):
                current_is_univalued = False
            else:
                current_size += left_size

            # Check if right subtree is univalued and matches current node's value
            if not is_right_univalued or (node.right and node.val != right_val):
                current_is_univalued = False
            else:
                current_size += right_size

            if current_is_univalued:
                self.max_size = max(self.max_size, current_size)
                return (True, current_size, node.val)
            else:
                return (False, 0, None) # This subtree is not univalued, its size doesn't count for the univalued property
                                         # The 0 and None here just indicate it's not univalued.
                                         # You could also return max(left_size, right_size) for the size,
                                         # but it's simpler to just set max_size globally.

        dfs(root)
        return self.max_size
```

9) String expression evaluation
Ex: 2a+2b-(a-b)
Ans: a+3b

```python
import collections

class Solution:
    def solveExpression(self, expression: str) -> str:
        # Dictionary to store coefficients of variables
        # e.g., {'a': 1, 'b': 3}
        coeffs = collections.defaultdict(int)

        # sign_stack: Stores the cumulative sign effect of nested parentheses.
        # Top of stack represents the current multiplier for any term inside current scope.
        # [1] means positive overall context. If we see '-(', we push -1 * current_top.
        sign_stack = [1] 
        
        # current_term_sign: The sign (+1 or -1) applied to the *next* number or variable encountered
        # within the current parenthetical scope. This is determined by '+' or '-'.
        current_term_sign = 1 
        
        i = 0
        n = len(expression)

        while i < n:
            char = expression[i]

            if char == ' ':
                i += 1
                continue

            elif char == '+':
                current_term_sign = 1
                i += 1
            
            elif char == '-':
                current_term_sign = -1
                i += 1

            elif char == '(':
                # When opening a parenthesis, the sign of its content is
                # current_term_sign * the overall sign of the parent scope.
                sign_stack.append(current_term_sign * sign_stack[-1])
                # After '(' reset current_term_sign for the terms *inside* the new scope
                current_term_sign = 1 
                i += 1

            elif char == ')':
                # When closing a parenthesis, pop the multiplier for that scope.
                sign_stack.pop()
                # After ')' reset current_term_sign for the terms *after* the closed scope
                # This depends on the problem specifics, usually it resets to +1 for the next term
                current_term_sign = 1 
                i += 1

            else: # Must be a number or a variable
                num_str = ""
                # Parse the number (coefficient)
                # If it starts with a digit, read all digits
                while i < n and expression[i].isdigit():
                    num_str += expression[i]
                    i += 1
                
                coefficient = int(num_str) if num_str else 1 # Default to 1 if no number specified (e.g., 'a' means 1a)
                
                # Parse the variable (if present)
                var_char = ''
                if i < n and 'a' <= expression[i] <= 'z':
                    var_char = expression[i]
                    i += 1
                
                # Calculate the final multiplier for this term
                # It's (sign for this term) * (overall sign from parent scopes)
                total_multiplier = current_term_sign * sign_stack[-1]
                
                # Apply the coefficient to the variable
                # If there's no variable (e.g., '5' without 'a'), we could handle it as a constant,
                # but the example '2a+2b-(a-b)' implies no standalone constants.
                # If constants were allowed, you'd use a special key in `coeffs` or a separate variable.
                if var_char:
                    coeffs[var_char] += coefficient * total_multiplier
                # else:
                #     # Handle standalone constants if necessary, e.g., coeffs['__constant__'] += coefficient * total_multiplier

        # --- Format the output string ---
        result_terms = []
        
        # Get sorted variables
        sorted_vars = sorted(coeffs.keys())

        for var in sorted_vars:
            coeff = coeffs[var]
            if coeff == 0:
                continue # Skip terms with zero coefficient

            term_str = ""
            if coeff > 0:
                # Add '+' only if it's not the very first term and positive
                if result_terms:
                    term_str += "+"
            elif coeff < 0:
                term_str += "-"
                coeff = abs(coeff) # Use absolute value for number part

            if coeff != 1 or not var: # Always print number if not 1 or if it's a constant
                if var and coeff == 1: # Special case for 1a to be 'a'
                    pass
                else:
                    term_str += str(coeff)

            term_str += var
            result_terms.append(term_str)
        
        if not result_terms:
            return "" # Or "0" if the problem implies a constant zero
        
        # Handle leading '+' if it's the first term
        final_result = "".join(result_terms)
        if final_result.startswith('+'):
            final_result = final_result[1:]

        return final_result

# Example Usage:
solver = Solution()
print(f"2a+2b-(a-b) -> {solver.solveExpression('2a+2b-(a-b)')}") # Expected: a+3b
print(f"a-b+(c-a+b) -> {solver.solveExpression('a-b+(c-a+b)')}") # Expected: c
print(f"-(a-b) -> {solver.solveExpression('-(a-b)')}")           # Expected: -a+b
print(f"a+b-a-b -> {solver.solveExpression('a+b-a-b')}")         # Expected: "" or "0" depending on spec, here ""
print(f"a -> {solver.solveExpression('a')}")                     # Expected: a
print(f"-a -> {solver.solveExpression('-a')}")                   # Expected: -a
print(f"5x - (3x + 2y) + 4y -> {solver.solveExpression('5x - (3x + 2y) + 4y')}") # Expected: 2x+2y
print(f"x-(y-(z-x)) -> {solver.solveExpression('x-(y-(z-x))')}") # Expected: -y+z
```

11) Given a node in a Binary Search Tree (BST), find the next largest (inorder successor) and next smallest (inorder predecessor) node. It is guaranteed that these nodes always exist (i.e the input node will never be the smallest or largest node in the BST).
You are only given a pointer to the node, and each node has a parent pointer.
The structure is as follows:
struct Node {
    int value;
    Node* left_child;
    Node* right_child;
    Node* parent;
};
Consider this BST:
        4
       / \
      2   6
     / \ / \
    1  3 5  7
If the given node is 3, the next largest (successor) is 4, the next smallest (predecessor) is 2.

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right= None
        self.parent = None

class Solution:
    def __init__(self):
        self.nodes = []
    def exploreTree(self, root):
        if not root:
            return
        self.exploreTree(root.left)
        self.nodes.append(root.val)
        self.exploreTree(root.right)

    def inOrderPredSucc(self, node):
        for i in range(len(self.nodes)):
            if self.nodes[i] == node.val:
                return (self.nodes[i-1], self.nodes[i+1])
        return 
    

def build_bst():
    nodes = {i: Node(i) for i in range(1, 8)}
    nodes[4].left = nodes[2]
    nodes[4].right = nodes[6]
    nodes[2].parent = nodes[4]
    nodes[6].parent = nodes[4]

    nodes[2].left = nodes[1]
    nodes[2].right = nodes[3]
    nodes[1].parent = nodes[2]
    nodes[3].parent = nodes[2]

    nodes[6].left = nodes[5]
    nodes[6].right = nodes[7]
    nodes[5].parent = nodes[6]
    nodes[7].parent = nodes[6]

    return nodes


def test():
    nodes = build_bst()
    root = nodes[4]
    sol = Solution()
    sol.exploreTree(root)

    test_node = nodes[5]
    pred, succ = sol.inOrderPredSucc(test_node)
    print(f"For node {test_node.val}, predecessor = {pred}, successor = {succ}")

test()
```

In a BST:

The inorder successor of a node is the smallest node in its right subtree, or the first ancestor for which the node is in the left subtree.

The inorder predecessor of a node is the largest node in its left subtree, or the first ancestor for which the node is in the right subtree.

And since each node has a parent pointer, we can walk up the tree without needing the root.

```python
class Solution:
    def inorder_successor(self, node):
        # Case 1: Right child exists → go down to the leftmost node
        if node.right:
            node = node.right
            while node.left:
                node = node.left
            return node

        # Case 2: No right child → go up until we come from the left
        while node.parent and node == node.parent.right:
            node = node.parent
        return node.parent

    def inorder_predecessor(self, node):
        # Case 1: Left child exists → go down to the rightmost node
        if node.left:
            node = node.left
            while node.right:
                node = node.right
            return node

        # Case 2: No left child → go up until we come from the right
        while node.parent and node == node.parent.left:
            node = node.parent
        return node.parent

def build_bst():
    nodes = {i: Node(i) for i in range(1, 8)}
    nodes[4].left = nodes[2]
    nodes[4].right = nodes[6]
    nodes[2].parent = nodes[4]
    nodes[6].parent = nodes[4]

    nodes[2].left = nodes[1]
    nodes[2].right = nodes[3]
    nodes[1].parent = nodes[2]
    nodes[3].parent = nodes[2]

    nodes[6].left = nodes[5]
    nodes[6].right = nodes[7]
    nodes[5].parent = nodes[6]
    nodes[7].parent = nodes[6]

    return nodes


def test():
    nodes = build_bst()
    root = nodes[4]
    sol = Solution()

    test_node = nodes[5]
    pred, succ = sol.inorder_predecessor(test_node).val, sol.inorder_successor(test_node).val
    print(f"For node {test_node.val}, predecessor = {pred}, successor = {succ}")

test()
```

13) Given 2 strings, return if they are buddies. 2 strings are buddies if the distance between each char of the strings is the same. dist('a', 'b') = 1 and dist('b', 'a') = 25. Also, dist('z', 'a') = 1.
Follow Up
Given a list of strings, group all buddies together.

```python
def dist(a,b):
    if ord(a) > ord(b):
        return 26 - (ord(a) - ord(b))
    else:
        return abs(ord(a) - ord(b))
    

def buddies(a,b):
    if len(a) != len(b): return False
    prev_distance = -1
    for c_a, c_b in zip(a, b):
        curr_dist = dist(c_a, c_b)
        if prev_distance != -1:
            if curr_dist != prev_distance:
                return False
        prev_distance = curr_dist

    return True

# buddies("apple", "bqqmf")

def buckets(input):
    res = []
    visited = set()
    n = len(input)
    for i in range(n):
        if input[i] in visited: continue
        curr = [input[i]]
        visited.add(input[i])
        for j in range(i+1, n):
            if input[j] in visited: continue
            if buddies(input[i], input[j]):
                curr.append(input[j])
                visited.add(input[j])

        res.append(curr)

    return res






input_strings = [
    "apple",   # len 5, Normalized: (15, 0, 22, 19)
    "bqqmf",   # len 5, Normalized: (15, 0, 22, 19) - Buddy of "apple" (constant shift +1)
    "cppmg",   # len 5, Normalized: (13, 0, 23, 20) - Not buddy of "apple"

    "fruit",   # len 5, Normalized: (12, 3, 14, 11)
    "gsvju",   # len 5, Normalized: (12, 3, 14, 11) - Buddy of "fruit" (constant shift +1)

    "cat",     # len 3, Normalized: (24, 19)
    "dog",     # len 3, Normalized: (11, 18) - Different from "cat"
    "fox",     # len 3, Normalized: (9, 9)
    "gpy",     # len 3, Normalized: (9, 9) - Buddy of "fox" (constant shift +1)

    "top",     # len 3, Normalized: (21, 1)

    "zzz",     # len 3, Normalized: (0, 0)
    "aaa",     # len 3, Normalized: (0, 0) - Buddy of "zzz" (constant shift variable, but internal is same)
    "bbb",     # len 3, Normalized: (0, 0) - Buddy of "zzz" and "aaa"

    "az",      # len 2, Normalized: (25)
    "ba",      # len 2, Normalized: (25) - Buddy of "az" (constant shift +1)

    "zz",      # len 2, Normalized: (0)
    "aa",      # len 2, Normalized: (0) - Buddy of "zz"

    "a",       # len 1, Normalized: ()
    "b",       # len 1, Normalized: () - Buddy of "a"
    "c",       # len 1, Normalized: () - Buddy of "a" and "b"
    "d",       # len 1, Normalized: ()

    "hello",   # len 5, Normalized: (21, 0, 15, 9) - Unique
    "world",   # len 5, Normalized: (8, 16, 17, 7) - Unique
    "",        # Empty string, Normalized: () - (If empty strings are allowed, typically they'd group together or be ignored)
    ""         # Another empty string
]

buckets(input_strings)
```

```python
#optimized

# Assuming the dist function is defined as previously confirmed correct:
def dist(a, b):
    # Calculates cyclic distance from char 'a' to char 'b'
    # e.g., dist('a', 'b') = 1, dist('b', 'a') = 25, dist('z', 'a') = 1
    return (ord(b) - ord(a) + 26) % 26

def normalize_string(s):
    """
    Computes a tuple representing the internal character differences for a string.
    This tuple serves as a unique key for grouping buddy strings.
    """
    if not s:
        return (-1) # Empty tuple for empty string, not buddies "" and single char

    # For a single character string, there are no internal differences
    if len(s) == 1:
        return () 

    differences = []
    for i in range(len(s) - 1):
        # Calculate cyclic distance between s[i] and s[i+1]
        diff = dist(s[i], s[i+1])
        differences.append(diff)
    
    return tuple(differences) # Return as a tuple for hashability

def optimized_buckets(input_strings):
    """
    Groups buddy strings together using a canonical form (internal differences).
    """
    groups = {} # Dictionary to store groups, key: canonical_form, value: list of strings

    for s in input_strings:
        # Step 1: Compute the canonical form for the current string
        canonical_form = normalize_string(s)
        
        # Step 2: Add the string to the list associated with its canonical form
        if canonical_form not in groups:
            groups[canonical_form] = []
        groups[canonical_form].append(s)
    
    # Step 3: Return all the lists of grouped strings
    return list(groups.values())

# Example usage with the previous test case:
input_strings = [
    "apple", "bqqmf", "cppmg", "fruit", "gsvju", "cat", "dog", "fox", "gpy", "top",
    "zzz", "aaa", "bbb", "az", "ba", "zz", "aa", "a", "b", "c", "d", "hello", "world", "", ""
]

# Note: The output order of groups and strings within groups is not guaranteed,
# but the content of each group will be correct.
grouped_buddies = optimized_buckets(input_strings)
for group in grouped_buddies:
    print(group)

# Example output would look something like this (order may vary):
# ['hello']
# ['bqqmf', 'apple']
# ['fox', 'gpy']
# ['ba', 'az']
# ['zzz', 'aaa', 'bbb']
# ['fruit', 'gsvju']
# ['d', 'a', 'b', 'c', '', ''] # Single char and empty strings group together
# ['world']
# ['top']
# ['zz', 'aa']
# ['dog']
# ['cat']
# ['cppmg']
```

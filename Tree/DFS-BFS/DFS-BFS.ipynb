# 572. Subtree of Another Tree

```python
#brute-force O(n*m) , O(n+m) space
class Solution:
    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        def _is_identical(rootA, rootB):
            queue = deque([(rootA, rootB)])
            while queue:
                a,b = queue.popleft()
                if a.val != b.val:
                    return False
                if a.left and b.left:
                    queue.append((a.left, b.left))
                elif (a.left and not b.left) or (not a.left and b.left):
                    return False
                if a.right and b.right:
                    queue.append((a.right, b.right))
                elif (a.right and not b.right) or (not a.right and b.right):
                    return False

            return True


        def _dfs(root, subRoot):
            if not root:
                return False

            if root.val == subRoot.val and _is_identical(root, subRoot):
                    return True

            return _dfs(root.left, subRoot) or _dfs(root.right, subRoot)





        return _dfs(root, subRoot)
    

#String matching, O(n+m) time, O(n+m) space
#KMP algorithm
#https://www.geeksforgeeks.org/kmp-algorithm-for-pattern-searching/
class Solution:
    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:

        # Function to serialize Tree
        def serialize(node, tree_str):
            if node is None:
                tree_str.append('#')
                return

            tree_str.append("^")
            tree_str.append(str(node.val))
            serialize(node.left, tree_str)
            serialize(node.right, tree_str)

        # Knuth-Morris-Pratt algorithm to check if `needle` is in `haystack`
        def kmp(needle, haystack):
            m = len(needle)
            n = len(haystack)

            if n < m:
                return False

            # longest proper prefix which is also suffix
            lps = [0]*m
            # Length of Longest Border for prefix before it.
            prev = 0
            # Iterating from index-1. lps[0] will always be 0
            i = 1

            while i < m:
                if needle[i] == needle[prev]:
                    # Length of Longest Border Increased
                    prev += 1
                    lps[i] = prev
                    i += 1
                else:
                    # Only empty border exist
                    if prev == 0:
                        lps[i] = 0
                        i += 1
                    # Try finding longest border for this i with reduced prev
                    else:
                        prev = lps[prev-1]

            # Pointer for haystack
            haystack_pointer = 0
            # Pointer for needle.
            # Also indicates number of characters matched in current window.
            needle_pointer = 0

            while haystack_pointer < n:
                if haystack[haystack_pointer] == needle[needle_pointer]:
                    # Matched Increment Both
                    needle_pointer += 1
                    haystack_pointer += 1
                    # All characters matched
                    if needle_pointer == m:
                        return True
                else:
                    if needle_pointer == 0:
                        # Zero Matched
                        haystack_pointer += 1
                    else:
                        # Optimally shift left needle_pointer. 
                        # Don't change haystack_pointer
                        needle_pointer = lps[needle_pointer-1]
            
            return False

        # Serialize given Nodes
        root_list = []
        serialize(root, root_list)
        r = "".join(root_list)

        subroot_list = []
        serialize(subRoot, subroot_list)
        s = "".join(subroot_list)

        # Check if s is in r
        return kmp(s, r)

# Tree hash O(n+m) time, O(n+m) space
class Solution:
    def isSubtree(self, root: TreeNode, subRoot: TreeNode) -> bool:

        MOD_1 = 1_000_000_007
        MOD_2 = 2_147_483_647

        def hash_subtree_at_node(node, need_to_add):
            if node is None:
                return (3, 7)

            left = hash_subtree_at_node(node.left, need_to_add)
            right = hash_subtree_at_node(node.right, need_to_add)

            left_1 = (left[0] << 5) % MOD_1
            right_1 = (right[0] << 1) % MOD_1
            left_2 = (left[1] << 7) % MOD_2
            right_2 = (right[1] << 1) % MOD_2

            hashpair = ((left_1 + right_1 + node.val) % MOD_1,
                        (left_2 + right_2 + node.val) % MOD_2)

            if need_to_add:
                memo.add(hashpair)

            return hashpair

        # List to store hashed value of each node.
        memo = set()

        # Calling and adding hash to List
        hash_subtree_at_node(root, True)

        # Storing hashed value of subRoot for comparison
        s = hash_subtree_at_node(subRoot, False)

        # Check if hash of subRoot is present in memo
        return s in memo
```

# leet 543

```python
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        res = 0

        def dfs(root):
            if not root:
                return 0
            
            l = dfs(root.left)
            r = dfs(root.right)

            nonlocal res
            res = max(res, l + r)

            return 1 + max(l, r)

        dfs(root)
        return res
```

# LOWEST COMMON ANCESTOR

```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        """
        LCA for BST
        """                
        while root:
            if p.val > root.val and q.val > root.val:
                root = root.right
            elif p.val < root.val and q.val < root.val:
                root = root.left
            else:
                return root
                
    def lca(root, p, q):
        """
        Return the lowest common ancestor of nodes p and q
        in an ordinary binary tree whose root is `root`.
        Nodes are compared by identity (not by value).
        """
        # Base‑cases
        if root is None or root is p or root is q:
            return root            # Found a target or hit the end

        # Recurse on children
        left  = lca(root.left,  p, q)
        right = lca(root.right, p, q)

        # If we got a hit from both sides, root is their first meeting point
        if left and right:
            return root

        # Otherwise propagate the non‑null result upward
        return left or right
```

# leet 1448

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def goodNodes(self, root: TreeNode) -> int:
        
        def count_good(root, curr_max):
            if not root:
                return 0

            if root.val > curr_max:
                curr_max = root.val
            l = count_good(root.left, curr_max)
            r = count_good(root.right, curr_max)
            if root.val >= curr_max:
                return 1 + l + r
            return l + r
        

        return count_good(root, root.val)
```

```python
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:

        def bst(root, low, high):
            if not root:
                return True

            if not (low< root.val < high):
                return False

            return bst(root.left, low, root.val) and bst(root.right, root.val, high)

        return bst(root, -float("inf"), float("inf"))
```

# leet 105
🔹 1. Preorder drives the root creation
Preorder is used sequentially to pick the next root node.

That’s why we keep a preorder_index that moves forward as we build the tree.

🔹 2. Inorder is used to split into left/right subtrees
Once we have a root value (from preorder), we find its position in the inorder list.

This splits the inorder segment into:

Left of the root → left subtree

Right of the root → right subtree

🔹 3. Recursive calls use the inorder subarray bounds
Each recursive call works only within a subrange of the inorder array.

If that subrange becomes invalid (i.e. left > right), it means:

📌 Base case → No more nodes to add → return None (leaf reached)




What	        Used for <br>
preorder	    Sequential root node selection<br>
inorder	        Left/right subtree boundary detection<br>
preorder_index	Tracks which root we’re building next<br>
Recursive       left > right	Indicates no more nodes (leaf reached)<br>


```python

# class TreeNode:
#     def __init__(self, val: int):
#         self.val = val
#         self.left: Optional['TreeNode'] = None
#         self.right: Optional['TreeNode'] = None
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        if not preorder or not inorder:
            return None

        #fast_lookup val -> index
        lookup = {val:i for i, val in enumerate(inorder)}
        preorder_index = 0

        def arr2tree(left, right):
            nonlocal preorder_index
            if left > right:
                return None

            root = TreeNode(preorder[preorder_index])
            preorder_index += 1

            root.left = arr2tree(left, lookup[root.val] - 1)
            root.right = arr2tree(lookup[root.val]+1, right)


            return root

        return arr2tree(0, len(inorder)-1)
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        #brute force, for each node explore the path sum
        max_val = 0
        if not root:
            return None
        max_res = root.val
        
        def sumpath(root):
            nonlocal max_res
            if not root:
                return 0

            l = max(sumpath(root.left), 0)
            r = max(sumpath(root.right), 0)
            max_res = max(max_res, l + root.val + r) #combine paths

            return root.val + max(l,r) #single path
        
        sumpath(root)
        return max_res
        #optimized = dfs
```
